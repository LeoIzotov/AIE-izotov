## Работа с NumPy

### 1. Импорт библиотеки

```python
import numpy as np
```

### 2. Создание массивов

```python
arr_from_list = np.array([1, 2, 3, 4, 5])
print("Массив из списка: ", arr_from_list)

zeros_arr = np.zeros(5)
print("Массив нулей: ", zeros_arr)

ones_arr = np.ones(5)
print("Массив единиц: ", ones_arr)

range_arr = np.arange(0, 12, 2)  # от 0 до 10 с шагом 2
print("Массив с помощью arange: ", range_arr)
```

### 3. Атрибуты массивов

```python
# Исследуем массив range_arr
print("Форма массива: ", range_arr.shape)
print("Тип данных массива: ", range_arr.dtype)
print("Размерность массива: ", range_arr.ndim)
```

### 4. Индексирование и срезы

```python
# Получение элементов с помощью обращения по индекву
print("Первый элемент: ", range_arr[0])
print("Последний элемент: ", range_arr[-1])

# Извлечение подмассивов с помощью срезов
print("Срез [2:5]: ", range_arr[2:5])  # Элементы с индексом 2 до 4 (5 не включительно)
print("Срез [:3]: ", range_arr[:3])    # Элементы от начала до индекса 2
```

### 5. Векторные операции и статистика

```python
A = np.array([1, 2, 3, 4, 5])
B = np.array([10, 20, 30, 40, 50])

# Векторные операции
print("Поэлементное сложение: ", A + B)
print("Поэлементное умножение: ", A * B)
print("Умножение на скаляр: ", A * 2)

# Базовые статистические функции
print("Среднее значение: ", np.mean(A))
print("Сумма элементов: ", np.sum(A))
print("Максимальный элемент: ", np.max(A))
print("Минимальный элемент: ", np.min(A))
print("Стандартное отклонение: ", np.std(A))
```

Формула стандартного отклонения через матожидание СВ X
$$\sigma = \sqrt{D(X)} = \sqrt{M\left[(X - M(X))^2\right]}$$
